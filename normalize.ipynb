```python
# This notebook is used to normalize the density of the species to its initial density
# (If the density of the species is unperturbed, it should equal to 1)
import numpy as np
import h5py
import os
import shutil
import glob
from scipy import interpolate
from helper import *

angle = 0

def get_file_list():
    file_list = glob.glob('../Species1/Charge/Merged_angle_' + str(angle) + '/*.h5')
    file_list = [os.path.basename(file) for file in file_list]
    return file_list

### Get the local density for each file
s, fs = get_density_profile(name = 'species', idx=0, plot=False, save=False, path = '..')
f = interpolate.interp1d(s, fs)

_,s_files = get_numbers_in_filenames(path = '..')

local_density = f(s_files)
###


### Normalize the plasma density
if not os.path.exists('../Species1/Charge/Merged_angle_' + str(angle) + '_normalized'):
    shutil.copytree('../Species1/Charge/Merged_angle_' + str(angle),\
                    '../Species1/Charge/Merged_angle_' + str(angle) + '_normalized')

    file_list = get_file_list()
    postfixs = [i.split('_')[1] for i in file_list]
    postfixs.sort()

    for i in range(len(postfixs)):
        postfix = postfixs[i]
        filename = '../Species1/Charge/Merged_angle_' + str(angle) + '_normalized' + '/charge_'+ postfix
        print('Normalizing' + str(filename) + '...')
        with h5py.File(filename,'r+') as h5file:
            # read dataset name and attributes
            dset_name = list(h5file.keys())[1]
            h5dset = h5file[dset_name]
            rho = np.array(h5dset)
            rho_normalized = rho / local_density[i]
            h5dset[...] = rho_normalized
### Normalize the Focusing force according to the local density
if os.path.exists('../Fields/Fr'):
    if not os.path.exists('../Fields/Fr/Merged_angle_' + str(angle) + '_normalized'):
        shutil.copytree('../Fields/Fr/Merged_angle_' + str(angle),\
                       '../Fields/Fr/Merged_angle_' + str(angle) + '_normalized')
        
        file_list = get_file_list()
        postfixs = [i.split('_')[1] for i in file_list]
        postfixs.sort()

        for i in range(len(postfixs)):
            postfix = postfixs[i]
            filename = '../Fields/Fr/Merged_angle_' + str(angle) + '_normalized' + '/fr_'+ postfix
            print('Normalizing' + str(filename) + '...')
            with h5py.File(filename,'r+') as h5file:
                # read dataset name and attributes
                dset_name = list(h5file.keys())[1]
                h5dset = h5file[dset_name]
                Fr = np.array(h5dset)
                Fr_normalized = Fr / local_density[i]
                h5dset[...] = Fr_normalized
```
